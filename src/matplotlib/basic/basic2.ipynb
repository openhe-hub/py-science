```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
import seaborn

plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
#setting
figure=plt.figure(figsize=[10,10])

ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')#设置标签位于x轴下方
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')#设置标签位于y轴左方
ax.spines['left'].set_position(('data',0))
x_ticks=np.arange(-5,5.5,1)
y_ticks=np.append(np.arange(-5,0,1),np.arange(0.5,5.5,1))
plt.xticks(x_ticks)
plt.yticks(y_ticks)#设置坐标轴标签
plt.axis('equal')#强制单位距离等长
text_offset=0.1




#generate data
a,b=2,1
zeta=np.arange(0,2*np.pi,0.01)
x1=[a*np.cos(i) for i in zeta]
y1=[b*np.sin(i) for i in zeta]
x2=np.arange(-3,3,0.01)
y2=[(i+1) for i in x2]
B1=[0,1]
A1=[-2,0]
F1=[-1.732,0]

#plot
plt.title('椭圆:'+r'$\frac{x^2}{4}+\frac{y^2}{1}=1$')
plt.plot(x1,y1,label=r'$\Gamma:\frac{x^2}{4}+\frac{y^2}{1}=1$')
plt.plot(x2,y2,label=r'$l:y=x+1$',linestyle=':')

plt.scatter(A1[0],A1[1],s=30,marker="x")
plt.text(A1[0]-text_offset,A1[1]+text_offset,'$A_1$',fontsize=8)
plt.scatter(B1[0],B1[1],s=15)
plt.text(B1[0]+text_offset,B1[1]+text_offset,'$B_1$',fontsize=8)
plt.scatter(F1[0],F1[1],s=15)
plt.text(F1[0]+text_offset,F1[1]+text_offset,'$F_1(-\sqrt{3},0)$',fontsize=8)

#display
plt.legend()
plt.show()
```

```python
def func(x):
    return (x - 3) * (x - 5) * (x - 7) + 85

def func1(x):
    return -x+60
x = np.linspace(0, 10)
y = func(x)
y1 = func1(x)


fig, ax = plt.subplots()
plt.plot(x, y, 'r', linewidth=2)
plt.plot(x, y1, 'g', linewidth=2)
plt.ylim(ymin=0)



a, b = 2, 9
xf = x[np.where((x>a)&(x<b))]

plt.fill_between(xf,func1(xf),func(xf),alpha=0.75)
plt.show()
```

```python
from scipy.optimize import minimize_scalar
def f(x):
    return -x**2+x;
min=minimize_scalar(f,bounds=(-2,2),method="bounded")
print(round(min.x,4))
```
